```python
from dash import Dash ,dash_table,dcc, html
import plotly.express as px
from dash.dependencies import Input,Output
import pandas as pd
import dash_bootstrap_components as dbc
import numpy as np
```

```python
app = Dash(external_stylesheets=[dbc.themes.BOOTSTRAP])


df = pd.read_csv("supermarket_sales - Sheet1.csv")
df2=pd.read_csv("grouped_data")

app.layout = html.Div([
    
    html.H1('Super Market Sales',style={"text-align":"center","color":"white"}),
    
    dbc.Row([
    dbc.Col(dbc.Card(dbc.CardBody([
                                   
       html.H2('Total Profit',style={"color":"white"}) ,
       html.H2('{} K'.format(np.round(df["gross income"].sum()/1000,2)),style={"color":"white"}) 
        
        
        
    ])#body
                  ,style={"background-color":"black"}  )#card
          , width = 4 , style={"background-color":"black"} )#col
    
    
    ,
        dbc.Col(dbc.Card(dbc.CardBody([
       html.H2("Total Saled Quantities",style={"color":"yellow"}) ,
       html.H2(df["Quantity"].sum(),style={"color":"yellow"}) 
        
        
        
    ])#body
                  , style={"background-color":"black"}  )#card
          , width = 4 , style={"background-color":"black"} )#col
        , dbc.Col(dbc.Card(
                           dbc.CardBody([
       html.H2('Average Rate',style={"color":"green"}) ,
       html.H2('{}/10'.format(np.round(df["Rating"].mean(),1)),style={"color":"green"}) 
        
        
        
    ])#body
                   , style={"background-color":"black"} )#card
          , width = 4 , style={"background-color":"black"} )#col
    ]
    
  , style={"background-color":"black"} ),#row
    
    html.Hr(),
    html.H4('Gender vs Product line',style={"text-align":"center","color":"white"}),
    html.P('select the month',style={"color":"white","margin":"8px"}),
    dbc.Row([
        dbc.Col(dcc.Dropdown(
        id="bar-chart-x-dropdown",
        options=[1,2,3],
        value=1,
        clearable=False,
    ),width=6)
    ]),
    


    dbc.Row([
    dbc.Col( dcc.Graph(id="bar-chart-x-graph",style={"margin":"8px"}),width=6),
    dbc.Col(dcc.Graph(figure=
             px.bar(df.groupby("Product line").sum()["gross income"].to_frame().sort_values("gross income"),
       title="Total profit from each product line",
       labels={'value':'Tolat profit'},template = 'plotly_dark').update_layout( bargap=0.5),style={"margin":"8px"})
     ,width=6)
]),
    html.Hr(),
    html.Hr(),
    html.H4('Some Analysis of the market sales',style={"text-align":"center","color":"white"}),
    html.P("Categories:",style={"color":"white","margin":"8px"}),
    dbc.Row([
        dbc.Col(dcc.Dropdown(id='names',
        options=['Branch', 'Gender', 'Payment'],
        value='Branch', clearable=False
    ),width=6)
    ]),    

    html.P("Value of:",style={"color":"white","margin":"8px"}),
    dbc.Row([
        dbc.Col(   dcc.Dropdown(id='values',
        options=['Total', 'Quantity', 'gross income',"Rating"],
        value='Total', clearable=False
    ),width=6)
    ]), 
    
    dbc.Row([
    dbc.Col( dcc.Graph(id="graph",style={"margin":"8px"}),width=6),
    dbc.Col(dcc.Graph(figure= px.histogram(df,x="day",y="Quantity",color="Gender"
                ,title="Sold quantities in days of the month",
                template = 'plotly_dark').update_layout( bargap=0.5),style={"margin":"8px"}),width=6)
]),

    html.Hr(),
    html.H4('Sold quantities of each product line in each branch',style={"color":"white","margin":"8px"}),
    dash_table.DataTable(
    data=df2.to_dict("records"),
    columns=[{'id': c, 'name': c} for c in df2.columns],
    page_action='none',
    style_table={'height': '600px', 'overflowY': 'auto'}
),
    
],style={"background-color":"black"})


@app.callback(
    Output("graph", "figure"), 
    Input("names", "value"), 
    Input("values", "value"))
def generate_chart(names, values):
    df = pd.read_csv("supermarket_sales - Sheet1.csv")
    fig = px.pie(df, values=values, names=names, hole=0.2,template = 'plotly_dark')
    return fig

@app.callback(
    Output("bar-chart-x-graph", "figure"), 
    Input("bar-chart-x-dropdown", "value"))
def update_bar_chart(m):
    df = pd.read_csv("supermarket_sales - Sheet1.csv") 
    mask = df["month"] == m
    fig = px.histogram(df[mask], x="Gender", y="gross income", 
                 color="Customer type", barmode="stack",title='Total profit by month of year',
                       template = 'plotly_dark').update_layout( bargap=0.5)
    return fig

app.run_server()
```
